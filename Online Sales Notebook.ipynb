```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt # standard
import csv
import seaborn as sns
from  itertools import groupby

sales = pd.read_csv("Online Sales Data.csv")
sales.describe()
```

```python
sales.head()
```

This is the basic relationship between depression and average grade as we see for every data

```python

#Creating a group by for average academic preformance for the filtered data
summary_data = sales.groupby('Product Category')['Total Revenue'].mean().reset_index()


plt.figure(figsize=(8, 6))
#making a bar chart to show the data in a presentation
plt.plot(summary_data['Product Category'], summary_data['Total Revenue'], color='red')

plt.title("Bar Plot of Average Total Revenue")
plt.xlabel("Type of Product")
plt.ylabel("Total Revenue")

plt.show()
```

#### Showing the same trend of the relationship between anxiety and depression increases as we add addtional stress factors


```python

#Creating a group by for average academic preformance for the filtered data
summary_data = sales.groupby('Payment Method')['Total Revenue'].sum().reset_index()


plt.figure(figsize=(8, 6))
#making a bar chart to show the data in a presentation
plt.plot(summary_data['Payment Method'], summary_data['Total Revenue'], color='red')

plt.title("Bar Plot of Total Revenue")
plt.xlabel("Type of Product")
plt.ylabel("Total Revenue")

plt.show()
```

showing that the data supports having 3 stressors that predict academic preformance cause the grades to drop below 2


```python

#Creating a group by for average academic preformance for the filtered data
summary_data = sales.groupby('Payment Method')['Units Sold'].sum().reset_index()


plt.figure(figsize=(8, 6))
#making a bar chart to show the data in a presentation
plt.plot(summary_data['Payment Method'], summary_data['Units Sold'], color='red')

plt.title("Bar Plot of Total Revenue")
plt.xlabel("Type of Product")
plt.ylabel("Total Revenue")

plt.show()

```

```python

sales["Date"] = pd.to_datetime(sales["Date"], errors='coerce')

sales["year_month"] = sales["Date"].dt.to_period('M')

summary_data = sales.groupby("year_month")["Units Sold"].sum().reset_index()


summary_data["year_month"] = summary_data["year_month"].astype(str)


plt.figure(figsize=(8, 6))
plt.bar(summary_data["year_month"], summary_data["Units Sold"], color='skyblue')


plt.title("Total Units Sold by Month")
plt.xlabel("Month")
plt.ylabel("Total Units Sold")


plt.xticks(rotation=45)

plt.show()
```

######  looking  at the columns of academic preformance we notice that the relationship academic preformance is moderatly strong negative or postive, so take any 3 moderatly strong relationship and will see a consentration of poor grades


```python

sales["Date"] = pd.to_datetime(sales["Date"], errors='coerce')

sales["year_month"] = sales["Date"].dt.to_period('M')

summary_data = sales.groupby("year_month")["Total Revenue"].sum().reset_index()


summary_data["year_month"] = summary_data["year_month"].astype(str)


plt.figure(figsize=(8, 6))
plt.bar(summary_data["year_month"], summary_data["Total Revenue"], color='skyblue')


plt.title("Total Revenue Sold by Month")
plt.xlabel("Month")
plt.ylabel("Total Revenue")


plt.xticks(rotation=45)

plt.show()
```
