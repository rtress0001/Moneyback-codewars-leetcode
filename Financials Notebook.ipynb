```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt # standard
import csv
import seaborn as sns
from  itertools import groupby

sales = pd.read_csv("Financials.csv")
sales.describe()
```

```python
sales.head()
sales.columns = (sales.columns
                 .str.strip()  # Remove leading/trailing whitespace
                 .str.lower()  # Convert to lowercase
                 .str.replace(' ', '_')  # Replace spaces with underscores
                )
sales.head()
```

This is the basic relationship between depression and average grade as we see for every data

```python
def convert_to_int(dollar_str):
    try:
        # Remove $ sign and commas, strip whitespace
        cleaned_str = dollar_str.replace('$', '').replace(',', '').strip()
        # Convert to float first to handle decimal values
        return int(float(cleaned_str))  # Convert float to int
    except ValueError:
        return None  # Return None if conversion fails

# Apply the conversion function to 'net_sales' and 'profit' columns
sales['net_sales'] = sales['gross_sales'].apply(convert_to_int)



#Creating a group by for average academic preformance for the filtered data
summary_data = sales.groupby('country')['net_sales'].sum().reset_index()

# Plot the bar chart
plt.figure(figsize=(8, 6))
plt.bar(summary_data['country'], summary_data['net_sales'], color='blue')

# Adding labels and title
plt.title("Average Gross Sales by Country", fontsize=14)
plt.xlabel("Country", fontsize=12)
plt.ylabel("Average Gross Sales", fontsize=12)


plt.show()
```

```python

sales["Date"] = pd.to_datetime(sales["Date"], errors='coerce')

sales["year_month"] = sales["Date"].dt.to_period('M')

summary_data = sales.groupby("year_month")["Units Sold"].sum().reset_index()


summary_data["year_month"] = summary_data["year_month"].astype(str)


plt.figure(figsize=(8, 6))
plt.bar(summary_data["year_month"], summary_data["Units Sold"], color='skyblue')


plt.title("Total Units Sold by Month")
plt.xlabel("Month")
plt.ylabel("Total Units Sold")


plt.xticks(rotation=45)

plt.show()
```

######  looking  at the columns of academic preformance we notice that the relationship academic preformance is moderatly strong negative or postive, so take any 3 moderatly strong relationship and will see a consentration of poor grades


```python

sales["Date"] = pd.to_datetime(sales["Date"], errors='coerce')

sales["year_month"] = sales["Date"].dt.to_period('M')

summary_data = sales.groupby("year_month")["Total Revenue"].sum().reset_index()


summary_data["year_month"] = summary_data["year_month"].astype(str)


plt.figure(figsize=(8, 6))
plt.bar(summary_data["year_month"], summary_data["Total Revenue"], color='skyblue')


plt.title("Total Revenue Sold by Month")
plt.xlabel("Month")
plt.ylabel("Total Revenue")


plt.xticks(rotation=45)

plt.show()
```
